# ISYE 4031 Student Onboarding

This short module helps you get comfortable with the tools used in the ISYE 4031 JupyterLite labs and homework notebooks. It is not a statistics module. It is a practical starter kit for running code, writing answers, saving your work, exporting PDFs, and recovering from browser storage issues.

If you are already comfortable with Jupyter notebooks, skim the first and last notebooks. If notebooks, Markdown, LaTeX, or Python data analysis are new to you, work through the sequence below before the first Python-heavy homework.


## Recommended order

The course files are listed in the file browser on the left side of the JupyterLite window. After you finish this notebook, double-click `01_jupyterlite_survival_guide.ipynb` in the left sidebar to open the next notebook.

1. `01_jupyterlite_survival_guide.ipynb`: open notebooks, run cells, save backups, export PDFs, and reset JupyterLite if needed.
2. `02_markdown_latex_quick_reference.ipynb`: write clear homework answers with Markdown and LaTeX.
3. `03_python_data_analysis_minimum.ipynb`: load data, inspect it, plot it, fit a simple model, and write a result.
4. `04_homework_submission_checklist.ipynb`: use this before submitting any homework.

You do not need to install Python to use the course JupyterLite sites. They run Python in your browser. If you choose to work in your own local Python environment instead, that is also fine as long as your submitted work answers the homework questions clearly.


## The one rule to remember

JupyterLite stores your edited notebooks in your browser's site storage. That means your work is usually still there when you come back on the same browser and device, but it is not the same as cloud storage.

Before clearing browser data, switching devices, or submitting a homework, download a backup copy of your `.ipynb` notebook and export the required PDF.


## Quick check

Open `01_jupyterlite_survival_guide.ipynb` from the left sidebar, run the image display cell, then try exporting that notebook to PDF. If the PDF contains the displayed image and the notebook text, your workflow is ready.
