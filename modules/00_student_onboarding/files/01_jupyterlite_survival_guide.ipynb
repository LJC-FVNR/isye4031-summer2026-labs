# JupyterLite Survival Guide

JupyterLite lets you run course notebooks in the browser. You can read explanations, run Python cells, edit Markdown answers, save your work, and export a PDF without installing a local Python environment.


## What is different about JupyterLite?

JupyterLite looks like JupyterLab, but it runs inside the browser. The Python kernel is provided by Pyodide, so the first package setup cell in a notebook can take a little time. After packages load, normal course workflows such as `pandas`, `matplotlib`, and `statsmodels` should feel familiar.

Your files are saved in browser storage for this JupyterLite site. They are not automatically synced to another browser, another computer, OneDrive, Google Drive, or Canvas.


## Run notebook cells

A notebook is a sequence of cells.

- A Markdown cell contains text, formulas, and explanations.
- A code cell contains Python code.
- To run the current cell, press `Shift+Enter` or click the run button.
- To run the whole notebook, use `Run > Run All Cells`.
- If the kernel gets confused, use `Kernel > Restart Kernel`, then run cells again from the top.

To edit a Markdown cell, double-click the text area of that cell. It will switch into editable Markdown source. After editing, run the cell to render it again. The Markdown rules are introduced in the next notebook.

To edit a code cell, click inside the cell and type directly. Run the cell when you want Python to execute the code.

For homework, run the notebook from top to bottom before exporting the final PDF. This catches missing imports, hidden state, and code cells that only worked because of an earlier experiment.


## Save and download a backup

Use both save and download. They solve different problems.

1. Save inside JupyterLite with `File > Save Notebook`.
2. Download a local backup from the left file browser. Right-click the notebook file and choose `Download`.
3. Keep that `.ipynb` file until the homework has been graded.

When working on homework in JupyterLite, save often and download a backup at least every 5-10 minutes. This greatly reduces the risk of losing work if the browser closes, the device sleeps, or site storage is unexpectedly cleared.

To continue from a downloaded backup later, use the upload button in the left file browser and upload the `.ipynb` file. If your JupyterLite browser storage is accidentally cleared while you are working, upload your most recent backup notebook into the left sidebar and continue from there.


## Export a PDF

A typical homework submission workflow is:

1. Run all cells from the top.
2. Save the notebook.
3. Use `File > Save and Export Notebook As > PDF`.
4. Open the exported PDF and check that written answers, formulas, code, plots, and images are visible.
5. Submit according to the Canvas instructions.

If PDF export fails, first save and download the `.ipynb` backup. Then try restarting the kernel, running all cells again, and exporting again.

## Reset or revert JupyterLite

If you want to return to the original course files, or if JupyterLite opens with stale files from an earlier attempt, clear the browser storage for this JupyterLite site.

Before clearing anything, download backup copies of notebooks you want to keep.

Preferred options inside JupyterLite:

1. Try `Help > Clear Browser Data` if it is available.
2. Or open the command palette and search for a clear browser data command.

Advanced browser fallback for Chrome or Edge:

1. Open the JupyterLite page.
2. Press `F12` to open Developer Tools.
3. Go to `Application > Storage`.
4. Choose `Clear site data` for the current site.
5. Reload the JupyterLite page.

After clearing site data, the published notebooks will appear in their original state again.
