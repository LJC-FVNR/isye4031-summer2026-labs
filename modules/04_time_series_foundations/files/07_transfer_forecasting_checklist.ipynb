```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.formula.api as smf
from statsmodels.stats.stattools import durbin_watson
from checks import check_columns, check_no_missing

plt.rcParams['figure.figsize'] = (8, 4)
plt.rcParams['axes.grid'] = True
bike = pd.read_csv(Path('data/trk50_bike_sales.csv'))
print(check_no_missing(bike))
bike.head()
```

## Checklist

1. Define the time unit and verify the observations are in the correct order.
2. Plot the response against time.
3. Look for trend: none, linear, quadratic, or another shape.
4. Look for seasonality: stable seasonal swing or increasing seasonal swing.
5. If the seasonal swing grows with the level, consider a transformation such as log.
6. Fit the simplest regression model that matches the visual structure.
7. Interpret coefficients on the correct scale and with the correct baseline season.
8. Check residuals over time and compute Durbin-Watson.
9. Use confidence intervals for mean response and prediction intervals for individual future observations.
10. Explain limitations, especially extrapolation and autocorrelation.

```python
response = 'Sales'
time_col = 'Time'
season_col = 'Quarter'

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(bike[time_col], bike[response], marker='o')
ax.set_xlabel(time_col)
ax.set_ylabel(response)
ax.set_title('Start every time-series regression with a plot')
plt.tight_layout()

model = smf.ols(f'{response} ~ {time_col} + C({season_col})', data=bike).fit()
print(model.summary().tables[1])
print(f'Durbin-Watson: {durbin_watson(model.resid):.3f}')
```

- Forecast: future settings, point estimates, confidence intervals, and prediction intervals.
