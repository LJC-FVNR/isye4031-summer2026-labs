# Homework Submission Checklist

Use this checklist before submitting a homework. The exact Canvas instructions for each assignment are the authority, but this page captures the repeated workflow.


## Your options

You may complete homework using the JupyterLite notebook provided for that homework, your own local Jupyter/Python environment, another appropriate computing environment, or handwritten work when the problem is a written problem.

The important requirement is that your submitted work clearly answers the questions, shows enough reasoning or code to justify the results, and follows the Canvas submission instructions.


## Before you start

- Read the full homework once before solving individual problems.
- Download any required data files if you are working outside JupyterLite.
- If using JupyterLite, open the homework notebook and save a personal working copy if you want to preserve the original.
- Keep formulas, code, and interpretation together. A numerical answer without context is often incomplete.


## While you work

- For written questions, use Markdown cells and LaTeX for formulas.
- For coding questions, run the provided setup/import cells first.
- Keep your code readable: clear variable names, short comments when helpful, and no unrelated scratch output in the final submission.
- After each major result, write a sentence interpreting the result in the language of the problem.
- If you are working in JupyterLite, save frequently and download a backup notebook at least every 5-10 minutes.


## AI Usage Report

Each homework requires an AI Usage Report according to the course policy. Follow the assignment and syllabus instructions.

A useful AI Usage Report should state:

- Whether generative AI tools were used.
- Which tool or tools were used.
- What the tool was used for, such as debugging code, explaining a concept, drafting wording, or checking algebra.
- What you did to verify the correctness of the final work.
- Which parts of the submitted work are your own final reasoning and edits.

If you did not use generative AI, say so clearly.


## Before exporting or submitting

- Add your identifying information if the assignment asks for it.
- Run the notebook from the top, preferably with `Run > Run All Cells`.
- Check that there are no missing outputs for required code cells.
- Check that Markdown formulas render correctly.
- Save the notebook.
- Download a `.ipynb` backup.
- Export the PDF.
- Open the PDF and verify that all written answers, formulas, code outputs, plots, and displayed images are visible.


## If something goes wrong

If code fails after reopening a notebook, restart the kernel and run all cells from the top.

If packages appear unavailable in JupyterLite, rerun the setup cell and wait for it to finish.

If the PDF is missing a plot or image, run the cell that creates or displays it, save the notebook, and export again. For images, prefer the `IPython.display.Image` method shown in the survival guide.

If your browser storage is unexpectedly cleared while you are working, upload your most recent downloaded `.ipynb` backup into the left file browser and continue from that backup.

If the site shows old files or a broken state, download backups first, then clear browser data for the JupyterLite site and reload the original link.


## Final quick checklist

- [ ] The submission includes the required homework answers.
- [ ] The submission includes any required code, plots, or numerical output.
- [ ] The submission includes the AI Usage Report.
- [ ] The PDF opens and is readable.
- [ ] A local `.ipynb` backup has been downloaded if you worked in JupyterLite.
- [ ] If working in JupyterLite, backups were downloaded during the work session, not only at the very end.
