# Markdown and LaTeX Quick Reference

Homework notebooks usually combine written explanation, formulas, code, and output. Markdown is used for written answers. LaTeX is used for math notation inside Markdown cells.


## Markdown basics

Use headings to organize answers. Markdown source:

```markdown
### Problem 1
### Interpretation
```

Rendered result:

### Problem 1
### Interpretation

Use lists for steps or assumptions. Markdown source:

```markdown
- Check whether the relationship is roughly linear.
- Fit the model.
- Interpret the slope in the units of the problem.
```

Rendered result:

- Check whether the relationship is roughly linear.
- Fit the model.
- Interpret the slope in the units of the problem.

Use backticks for code names or variable names. Markdown source:

```markdown
Use `sales`, `visitors`, and `model.summary()` in the answer.
```

Rendered result:

Use `sales`, `visitors`, and `model.summary()` in the answer.


## Links and compact tables

A link looks like this in Markdown source:

```markdown
[Course JupyterLite labs](https://ljc-fvnr.github.io/isye4031-summer2026-labs/)
```

Rendered result:

[Course JupyterLite labs](https://ljc-fvnr.github.io/isye4031-summer2026-labs/)

A small Markdown table looks like this in Markdown source:

```markdown
| Quantity | Meaning |
| --- | --- |
| slope | Estimated change in the response for a one-unit increase in x |
| p-value | Evidence against the null hypothesis under the model assumptions |
```

Rendered result:

| Quantity | Meaning |
| --- | --- |
| slope | Estimated change in the response for a one-unit increase in x |
| p-value | Evidence against the null hypothesis under the model assumptions |

Use tables sparingly in homework. A clear sentence is often better than a large table.


## Inline math

Use one dollar sign on each side for inline math.

```markdown
The fitted value is $\hat{y}_i = b_0 + b_1 x_i$.
```

Rendered result:

The fitted value is $\hat{y}_i = b_0 + b_1 x_i$.


## Display math

Use two dollar signs on each side for a centered formula.

```markdown
$$
\hat{\beta}_1 = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^n (x_i - \bar{x})^2}
$$
```

Rendered result:

$$
\hat{\beta}_1 = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^n (x_i - \bar{x})^2}
$$


## Common LaTeX patterns

| Goal | LaTeX |
| --- | --- |
| Subscript | `$x_i$` |
| Superscript | `$R^2$` |
| Fraction | `$\frac{a}{b}$` |
| Square root | `$\sqrt{x}$` |
| Average | `$\bar{x}$` |
| Estimate | `$\hat{\beta}_1$` |
| Sum | `$\sum_{i=1}^n x_i$` |
| Null hypothesis | `$H_0: \beta_1 = 0$` |
| Alternative hypothesis | `$H_a: \beta_1 \ne 0$` |
| Less than or equal | `$x \le 10$` |
| Greater than or equal | `$x \ge 10$` |


## Writing a good statistical answer

A useful homework answer usually has four parts:

1. The method or quantity you used.
2. The numerical result, with units when relevant.
3. The interpretation in the language of the problem.
4. Any important assumption or limitation.

Example:

The estimated slope is $b_1 = 0.42$. In this fitted model, an additional visitor is associated with an estimated increase of 0.42 sales units on average, holding the promotion indicator fixed. This interpretation is conditional on the fitted linear model being a reasonable summary of the data.
