# 00. Forecasting Context

Regression and forecasting are tools for making disciplined predictions from data. In this course, a prediction is not just a number. A useful prediction also tells us what information was used, what uncertainty remains, and what assumptions must be reasonable for the prediction to be trusted.

## Forecasts and forecasting

A **forecast** is a prediction of a future event, future value, or future condition. **Forecasting** is the process used to produce that prediction.

Forecasts matter because decisions often have to be made before the outcome is known:

- a production team estimates demand before scheduling labor,
- a finance team estimates future revenue before setting budgets,
- an energy provider estimates fuel needs before purchasing supply,
- an operations team estimates service volume before assigning capacity.

Most forecasting methods follow the same basic idea:

1. collect relevant historical data,
2. look for a pattern,
3. build a model that captures the useful part of the pattern,
4. extend the pattern to a future or unobserved case,
5. report uncertainty and limitations.

The dangerous phrase is "assuming the pattern continues." A model can only be useful when that assumption is reasonable enough for the decision at hand.

## Cross-sectional and time-series data

Two common data structures appear throughout the course.

**Cross-sectional data** contain several units observed at roughly the same point in time. One row might be a city, a customer, a product, a county, or a student.

Example: for many cities on one day, record maximum temperature, humidity, wind speed, and electricity demand.

**Time-series data** contain observations ordered over time, usually for the same process or unit. One row might be a day, week, month, quarter, or year.

Example: for one organization, record monthly revenue for five years.

The distinction matters because time-series observations can have trend, seasonal patterns, cycles, and serial dependence. Cross-sectional observations may instead emphasize differences across units.

## Components of a time series

When we study forecasting later in the course, we will often decompose a time series into:

- **trend**: long-run upward or downward movement,
- **cycle**: multi-period movement around a trend,
- **seasonality**: a repeated pattern within a calendar cycle,
- **irregular fluctuation**: movement not captured by regular structure.

Regression can still be used with time-indexed data, but we must be more careful about independence and time order.

## Where regression fits

Regression analysis predicts or explains a response variable using one or more predictors.

In simple linear regression, we use one predictor:

$$
Y = \beta_0 + \beta_1 x + \epsilon.
$$

In multiple regression, we use several predictors:

$$
Y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_p x_p + \epsilon.
$$

Regression is a forecasting method when the fitted model is used to predict future or unobserved values. It is also an explanatory method when the goal is to understand how the response changes with predictors.

Before modeling, always identify the unit of observation, the response variable, the predictor variables, and whether time order matters. This one habit prevents many modeling mistakes later.
